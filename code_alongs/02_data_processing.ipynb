```python

df["rating_count"] = df["rating_count"].str.replace(",", "").astype(float)
df["rating"] = df["rating"].str.replace("|","0").astype(float)
#df["rating"].astype(float)
df.head()
```

```python
df_pris = df[["product_name", "actual_price", "rating"]].sort_values(by=["actual_price"] ,ascending=False)
df_pris
```

```python
df_pris.drop(df_pris[df_pris["rating"]== 0].index, inplace=True)
```
